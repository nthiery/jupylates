```python
import numpy as np
import matplotlib.pyplot as plt

# Dfinition de la fonction complexe
def f_complex(x):
    return np.where(x != 0, x * (x**2+1) * np.cos(x) * np.sin(1 / x), 0)

# Gnration des valeurs de x, excluant 0 pour viter la division par zro
x = np.linspace(-5, 5, 10000)
x = x[x != 0]  # Exclusion de 0
y = f_complex(x)

# Cration de la figure
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='purple')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
